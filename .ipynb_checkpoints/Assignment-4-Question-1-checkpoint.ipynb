# Q1: In the netflix dataset replace all null values present in country column with united states.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_data.csv")
```

```python
data.head(20)
```

```python
data["country"] = data["country"].replace(np.nan, "United States")
```
